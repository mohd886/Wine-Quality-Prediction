# Random Forest Classifier

```python
param = {  'max_depth':[3,5,10,15,20,25],
           'n_estimators':[10,100,200,300,400,500],
           'max_features':['sqrt','log2','auto'],
           'criterion':['gini','entropy'],
           'bootstrap':[True,False],
           'min_samples_leaf': [1,2,3]
              }
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
clf = RandomForestClassifier(random_state=42)
```

```python
def hypertuning_rscv(classifier, parameters, epochs, X, y):
    random_search = RandomizedSearchCV(estimator = classifier,
                           param_distributions = parameters,
                           scoring = 'accuracy',
                           cv =5,
                           n_iter = epochs,
                           n_jobs = -1,
                           random_state=42)
    random_search.fit(X, y)
    ht_param = random_search.best_params_
    ht_score = random_search.best_score_
    return ht_param, ht_score
```

```python
ht_param, ht_score = hypertuning_rscv(clf, param, 100, X_train, y_train)
```

```python
ht_score
```

```python
ht_param
```

```python
clf = RandomForestClassifier( n_estimators = 400,
                                min_samples_leaf = 1,
                                max_features = 'auto',
                                max_depth = 20,
                                criterion ='gini',
                                bootstrap = False)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm_2 = confusion_matrix(y_test, y_pred)
```
